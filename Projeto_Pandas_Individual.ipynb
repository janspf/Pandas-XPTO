```python
import pandas as pd
```

```python
import numpy as np
```

```python
# utilização de base de dados
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQI152IpgRhcBTVWmGXP-oEQy4MNRSHJIQ29nocdEdOtNIpSE9ENGKUD3JeL6YHoA/pub?gid=1469769408&single=true&output=csv')
df.head()
```

```python
# calculando medidas de estatística descritiva para os atributos
descricao = df.describe()
print(descricao)
```

```python
# distribuição de frequência para atributos quantitativos
df['idade_faixa'] = pd.cut(df['Idade'], bins=5)  # Agrupando a idade em 5 faixas
distribuicao_idade = df['idade_faixa'].value_counts().sort_index()
print("Distribuição de Frequência para Idade:")
print(distribuicao_idade)

# distribuição de frequência para atributos qualitativos
distribuicao_departamento = df['Departamento'].value_counts()
print("\nDistribuição de Frequência para Departamento:")
print(distribuicao_departamento)

distribuicao_cargo = df['Cargo'].value_counts()
print("\nDistribuição de Frequência para Cargo:")
print(distribuicao_cargo)
```
